```python
def new_column(s1, df):
    lst = []
    for s in df['Amenities']:
        if s1 in s:
            lst.append('1')
        else:
            lst.append('0')
    return lst
    
def prop_type(df):
    lst = []
    for s in df['Property Type']:
        if 'Apartment' or 'Condominium' or 'Loft' in s:
            lst.append('1')
        elif 'House' or 'Townhouse' in s:
            lst.append('0')
        else:
            lst.append('')
    return lst

def cancel_policy(df):
    lst = []
    for s in df['Cancellation Policy']:
        if 'strict' in s:
            lst.append('2')
        elif 'moderate' in s:
            lst.append('0')
        elif 'flexible' in s:
            lst.append('0')
        else:
            lst.append('2')
    return lst
    
```

```python
X = dataset[:,0:12]
```

```python
y = dataset[:,12]
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
y=np.reshape(y, (-1,1))
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
print(scaler_x.fit(X))
xscale=scaler_x.transform(X)
print(scaler_y.fit(y))
yscale=scaler_y.transform(y)
```

```python
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(X)
y = StandardScaler().fit_transform(y.reshape(len(y),1))[:,0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(xscale, yscale)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python


model = Sequential()
model.add(Dense(12, input_dim=12, kernel_initializer='normal', activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['mse','mae'])
```

```python
history = model.fit(X_train, y_train, epochs=150, batch_size=50,  verbose=1, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
print(history.history.keys())
# "Loss"
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
Xnew = np.array([[10453.0, 1000.0, 2, 1, 97, 1, 0, 1, 1, 1, 1, 1]])
Xnew= scaler_x.transform(Xnew)
ynew= model.predict(Xnew)
#invert normalize
ynew = scaler_y.inverse_transform(ynew) 
Xnew = scaler_x.inverse_transform(Xnew)
print("X=%s, Predicted=%s" % (Xnew[0], ynew[0]))
```

```python
model.evaluate(X_test, y_test)
```
